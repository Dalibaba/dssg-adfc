```python
import pandas as pd
import pathlib
import os
```

```python
current_path = pathlib.Path().resolve()
BASE_DIR = os.path.join(current_path, "data" ) 
files = os.listdir(BASE_DIR)
```

```python
for file in files:
    file_dir = os.path.join(BASE_DIR, file)
    df = pd.read_csv(file_dir, index_col = 0)
    print(df)
    
```
